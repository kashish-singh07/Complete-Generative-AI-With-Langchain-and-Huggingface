Data Types

Python has several built-in data types to categorize data, which determines how values are stored and the operations that can be performed on them. Python is dynamically typed, so you do not need to explicitly declare the type of a variable; the interpreter determines it at runtime based on the assigned value.

The four most commonly used basic data types in Python are integers, floating-point numbers, strings, and booleans. 

Here is a breakdown of these and other key data types:

Numeric Types:

Used for storing numeric values.

    Integer (int): Represents whole numbers, positive or negative, without decimals (e.g., 10, -5).

    Floating-point number (float): Represents numbers with a decimal point (e.g., 3.14, 2.5).

    Complex number (complex): Represents numbers with a real and an imaginary part (e.g., 2 + 3j).


Sequence Types: 

Ordered collections of items.

    String (str): An immutable sequence of characters used for text (e.g., "Hello World").

    List (list): A mutable (changeable) ordered collection of items, which can be of different data types (e.g., ["apple", 1, True]).

    Tuple (tuple): An immutable ordered collection of items (e.g., (10, 20, 30)).


Boolean Type: 

    Represents truth values, used in conditional statements and decision-making.

    Boolean (bool): Has only two values: True or False.

    
Set Type: 

An unordered collection of unique items.

    Set (set): A mutable collection with no duplicate elements (e.g., {1, 2, 3}).

    Mapping Type: Stores data in key-value pairs.
    
    Dictionary (dict): An unordered collection that uses unique keys to access values (e.g., {"name": "John", "age": 30})
