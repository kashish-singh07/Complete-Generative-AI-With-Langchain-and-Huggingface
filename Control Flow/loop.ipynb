Loops

In Python, the two primary types of loops are for loops and while loops. These constructs are used to execute a block of code repeatedly to automate tasks and process data efficiently. 


1. for Loop

A for loop is used to iterate over a sequence (such as a list, tuple, string, or range) or other iterable objects. It is ideal when you know the number of iterations required beforehand

```python
##syntax
for variable in sequence:
    # Code block to execute for each item
```

```python
##Example
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
##range() function: Often used with for loops to loop a specific number of times. range(5) generates numbers from 0 to 4.
for i in range(5):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
```

    2. while Loop

A while loop executes a block of code repeatedly as long as a given condition remains True. It is best suited for situations where the number of iterations is not known in advance, and the loop's termination depends on a specific condition that can change during execution. 

```python
##syntax
while condition:
    # Code block to execute
```

```python
##Example
i = 1
while i < 6:
    print(i)
    i += 1 # Important: ensures the condition eventually becomes False
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

3. Nested Loops

Python allows you to use one loop inside the body of another loop, which is called a nested loop. This is useful for working with multi-dimensional data structures, like a list of lists or a 2D array. 
Loop Control Statements
You can manage the flow of loops using special control statements: 

break: 
Immediately terminates the current loop and transfers execution to the statement immediately following the loop.

continue: 
Skips the rest of the current iteration and returns control to the beginning of the loop for the next iteration.

pass: 
A null operation used as a placeholder when a statement is syntactically required but you don't want any code to execute. 

```python
##break example
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##continue example
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
##pass example
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
##nested loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
