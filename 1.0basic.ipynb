Syntax and Semantics

Sytax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. 
In simple words, syntax is about correct arrangement of words and symbols in code.

Semantics refers to the meaning or the interpretaton of the symbols, characters and commands in a language.It is about what the code is supposed to do  when it runs..

```python
##Basic syntax rules
##Case sensitivety
name="krish"
Name="naik"
print(name)
print(Name)
```
