Variable

In Python, a variable is a symbolic name (or label) that refers to a value or object stored in the computer's memory. Python is dynamically typed, which means you do not need to declare the variable's type; it is automatically determined when you assign a value to it using the assignment operator (=).

```python
a=100
```

```python
##declaring and assinging variaables
age=23
height=5.3
name="Kashish"
is_student=True
print("age:", age)
print("Height:", height)
print("Name:", name)
```

```python
##naming convention
##variables name should be descriptive
##they must start with a letter or '_' and contain letter, numbers, underscore
##variable name is case sensitive
##valid variable name

first_name="Kashish"
last_name="Singh"
```

```python
##variable type
age=23
height=5.3
print(type(height))
print(type(age))
```

```python
##type checking and conversion

age=23
print(type(age))
age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
age=25
print(type(int(age)))
```

```python
##dynamic typing
var = 100
print(var, type(var))
var="Honey"
print(var, type(var))
var=11.76
print(var, type(var))
```
