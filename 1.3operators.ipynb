Operators

Python operators are special symbols and keywords used to perform operations on variables and values, serving as fundamental tools for data manipulation and control flow. 

These operators are categorized based on their functionality: 

Arithmetic Operators

 (+, -, *, /, //, %, **) handle mathematical calculations, including standard arithmetic, floor division, and exponentiation.

Assignment Operators 

(=, +=, -=, etc.) assign values to variables, with compound options allowing for shorthand operations.

Comparison Operators
 (==, !=, >, <, >=, <=) evaluate relationships between values, returning True or False.

Logical Operators 

(and, or, not) are used to combine conditional statements.

Identity Operators

 (is, is not) check if two variables refer to the same object in memory.

Membership Operators 

(in, not in) test for the presence of a value within a sequence.

Bitwise Operators 

(&, |, ^, ~, <<, >>) operate on operands at the binary level. 


```python
##arithematic operation
a=10
b=5
result=a+b
floor_division=a//b
exponent_result=a**b
print(floor_division)
print(exponent_result)
print(result)
```
